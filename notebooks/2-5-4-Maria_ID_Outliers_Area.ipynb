```python
test=df.groupby(["theClass"])["area"].describe()
```

```python
Q1 = test["25%"]
Q3= test["75%"]

test["IQR"] = Q3 - Q1 
test["outliers_low"]=Q1 - 1.5 * test["IQR"]
test["outliers_high"]=Q3 + 1.5 * test["IQR"]
```

```python
test
```

```python
import os
import shutil

for i in range(0,38):
    df_class=df[df["theClass"]==test.index[i]]
    ans=df_class[(df_class["area"]<=test.iloc[i]["outliers_low"]) | (df_class["area"]>=test.iloc[i]["outliers_high"])]["name"]


    for j in ans:
        pic=j.split("\\")[-1]
        type_data=j.split("\\")[-3]

        destination = "c:\\Users\\Maria PaulaValentina\\Desktop\\DataScience\\Proyecto\\Outliers\\"+type_data+"\\"+test.index[i]
        # if os.path.isdir(destination):
        #     shutil.rmtree(destination)
        try:
            os.mkdir(destination)
        except:
            print("a")
        

        shutil.copy(j, destination+"\\"+pic)
```

```python
destination = "c:\\Users\\Maria PaulaValentina\\Desktop\\DataScience\\Proyecto\\Outliers_cleaned"
type_data=["train","valid"]

outliers_path=[]

for i in type_data:

    folderType = destination +"\\"+i

    theClass=os.listdir(folderType)

    for k in theClass:

        pic=os.listdir(folderType+"\\"+k)

        for l in pic:

            path=i+"\\"+k+"\\"+l
            outliers_path.append(path)



```
